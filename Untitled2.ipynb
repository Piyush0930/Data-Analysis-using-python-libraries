## Analyze the ratings and popularity of different restaurant chains.

```python
df.head()
```

```python
chain_summary = df.groupby('Restaurant Name').agg({
    'Aggregate rating': 'mean',
    'Restaurant ID': 'mean'
}).reset_index()
```

```python
chain_summary
```

```python
chain_summary_sorted_by_rating = chain_summary.sort_values(by='Aggregate rating', ascending=False)
chain_summary_sorted_by_popularity = chain_summary.sort_values(by='Restaurant ID', ascending=False)
```
