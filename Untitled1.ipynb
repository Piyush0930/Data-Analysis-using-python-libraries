### Compare the average ratings of restaurantswith and without online delivery

```python
df['online_delivery'] = df['Has Online delivery'].str.lower() == 'yes'
```

```python
average_ratings_with_delivery = df[df['online_delivery']]['Aggregate rating'].mean()
average_ratings_without_delivery = df[~df['online_delivery']]['Aggregate rating'].mean()
```
